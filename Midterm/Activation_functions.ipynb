```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(-10, 11, 1)
print(x)
def plot_graph(y,ylabel):
    plt.figure()
    plt.plot(x,y, 'o--')
    plt.grid(True)
    plt.xlabel("X")
    plt.ylabel(ylabel)
    plt.show()
```

<h5>Step Function</h5>
<p>A step function outputs either 1 if the input is above or, 0 if the input is below a certain threshold.</p>

```python
y = list(map(lambda n: 1 if n>0.5 else 0, x))
plot_graph(y, "STEP(X)")
```

<h5>Sigmoid or Logistic Activation Function</h5>
<p>The sigmoid function maps any input value to a range between 0 and 1. Therefore, it is especially used for models where we have to predict the probability as an output since probability of anything exists only between the range of 0 and 1</p>

```python
y = 1/(1 + np.exp(-x))
plot_graph(y, "Sigmoid(X)")
```

<h5>Tanh</h5>
<p>Like the sigmoid function, its activations saturate, but unlike the sigmoid function its output is zero-centered and outputs values in the range between -1 and 1</p>

```python
y = (np.exp(2*x) - 1) / (np.exp(2*x) + 1)
plot_graph(y, "tanh(X)")
```

<h5>Rectified Linear Units (ReLU)</h5>
<p>The activation is simply thresholded at zero. if the input x is positive, the function returns the same value, and if the input x is negative, the function returns zero.</p>

```python
y = list(map(lambda a: a if a>=0 else 0, x))
plot_graph(y, "ReLU(X)")
```

<h5>Exponential Linear Unit (ELU)</h5>
<p>The ELU function is introduced as an alternative to the ReLU function which has a hyperparameter alpha. It has a similar shape to the ReLU function for non-negative input values, whereas there is a smoother exponential transition for negative input values which helps in reducing the likelihood of dead neurons.</p>

```python
def Elu(x, alpha=0.6):
    return list(map(lambda a: a if a>=0 else alpha*(np.exp(a) - 1), x))
y = Elu(x)
plot_graph(y, "ELU(X)")
```

<h5>Scaled Exponential Linear Unit (SeLU)</h5>
<p>SeLU is introduced as a self-normalizing activation function, which has two hyperparameters alpha and scale that control the slope and scaling of the function for negative inputs. It helps for vanishing and exploding gradient problem that can occur in deep neural networks.</p>

```python
def SeLU(x, scale=1.0507, alpha=1.6733):
    return list(map(lambda a: scale*(a if a >= 0 else alpha*(np.exp(a) - 1)), x))
y = SeLU(x)
plot_graph(y, "SeLU(X)")
```
